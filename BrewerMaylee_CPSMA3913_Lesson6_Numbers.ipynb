So since 37 is negitive we have to add one to the reverse. Then we would get 011010+1=101101. Now for some code

```python
from math import floor
def binaryconvertor(x):
  ans = ''
  while x != 0:
    ans = str(x%2) + ans
    x = floor(x/2)
  return ans
binaryconvertor(25)
```

```python
def binaryconvertor(x):
  ans = ''
  if x<0:
    x = abs(x)
    ans = 'neg'
  while x != 0:
    ans = str(x%2) + ans
    x = floor(x/2)
  return ans
binaryconvertor(-37)
```

```python
def addone(list7binary):
  counter = 7
  list7binary[counter] = str(1 + int(list7binary[counter]))
  while list7binary[counter]=='2':
    list7binary[counter] = '0'
    counter -= 1
    list7binary[counter] = str(1 + int(list7binary[counter]))
  return list7binary

addone(['0','1','1','1','1','1','1','1'])
```

Now we'll do the reverse. 

```python
def flipit(list7binary):
  newlist = []
  for i in list7binary:
    if i == '0':
      newlist.append('1')
    else:
      newlist.append('0')
  return newlist

flipit(['0','1','1','1','1','1','1','1'])
```

```python

def twoscomplement(x):
  if not (x>-128)&(x<127):
    return "No Two's Complement"
  ans = '00000000'
  ans = list(ans) #strings don't allow you to change them in place (immutable) but lists can be!
  counter = 7
  if x>=0:
    while x != 0:
      ans[counter] = str(x%2)
      x = floor(x/2)
      counter -= 1
  if x<0:
    x = abs(x)
    while x != 0:
      ans[counter] = str(x%2)
      x = floor(x/2)
      counter -= 1
    ans = flipit(ans)
    addone(ans)        
  ans = ''.join(ans)
  return ans
```

```python
twoscomplement(-37)
```
