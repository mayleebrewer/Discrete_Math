# Compound event 

For my comound event we are going to see what the probability is for spinning two blues in a row. Since this will be equally likely on each of the two spins this should give us 4*4=16 possible outcomes. 

```
p= [(1/4)*(1/4)]
sum(p)
```

So there is a 0.0625 chance that we will spin a blue twice in a row!

## Union and intersection

This next part was a little tricky. We are going to split the spinner in half and say that the right side is blue and red and the left side is yellow and green. Now let's say that 

P(B)= the probability of the spinner landing on blue 

P(R)= the probability of the spinner landing on the right side 

P(B|R)= the probability of B given R

P(R|B)= the Probabilty of R given B

P(B∩R)= the probability of both. 

```
#P(B\capR)
I={(1/4)*(1/2)}
sum(I)
```

```
#P(B|R)
R=[((1/2)*(1/4))/(1/2)]
sum(R)
```

```
#P(R|B)
B=[((1/4)*(1/2))/(1/4)]
sum(B)
```

```
#Bayes Theorem P(B|P)
T={((1/4)*(0.5)) /(1/2)}
sum(T)
```

As you can see bayes theorem  held up!
